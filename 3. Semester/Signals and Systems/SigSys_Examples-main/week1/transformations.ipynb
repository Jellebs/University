## Define transforms

```python
def time_shift(ts, shift):
    return ts + shift
    
def time_scale(ts, scalar):
    return ts * scalar

def time_rev(ts):
    return time_scale(ts, -1)
```

## Transform

### Perform transform

```python
# shift
shift_with = 1
ys_shifted = signal_func(time_shift(ts, shift_with))

# reverse and shift
shift_with = 1
ys_rev_shift = signal_func(time_shift(time_rev(ts), shift_with))

# scale
scale_by = 3/2
ys_scaled = signal_func(time_scale(ts, scale_by))

# scaled and shift
scale_by = 3/2
shift_with = 1
ys_scale_shift = signal_func(time_shift(time_scale(ts, scale_by), shift_with))

x = ts
# nicely package signals into dictionary
transforms_dict = {
    'Shifted': ys_shifted,
    'Reversed and shifted': ys_rev_shift,
    'Scaled': ys_scaled,
    'Scaled and shifted': ys_scale_shift, 
    'x': x,
}
```

### Plot

```python
h.plot_transforms(ts, ys, transforms_dict)
```
