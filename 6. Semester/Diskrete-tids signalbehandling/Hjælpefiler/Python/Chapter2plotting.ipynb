```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

```

```python
# MATLAB: n=(-10:10); x=2*cos(2*pi*0.05*n)

n=np.arange(-10,11)
x=2*np.cos(2*np.pi*0.05*n)


fig, axs=plt.subplots(1,2)
axs[0].plot(n,x)
axs[1].stem(n,x)



```

```python
# numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)

# example: 1 second of samples, fs=31250 Hz

t=np.linspace(0,1,31250, endpoint=False)
print(t[0:5], t[-6:-1])
print(t.shape)

x=np.cos(2*np.pi*50*t)

fig, ax=plt.subplots()
ax.plot(t,x)
```

```python
a=x @ x.T
b=x.T @ x
c=np.dot(x,x)
d=np.inner(x,x)
print(a,b,c,d)
```
